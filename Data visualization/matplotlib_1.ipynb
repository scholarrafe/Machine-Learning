```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
y = x**2
plt.plot(x,y, linestyle='--', color='r', linewidth=5)
```

```python
fig1 = plt.figure(figsize=(8, 5))
ax = fig1.add_axes([0,0,1,1])
ax.plot(x,y, linewidth=5, color='r')
ax.set_title('Object Oriented plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
fig1.savefig("graphsave.png", dpi=200)
```

```python
fig, ax = plt.subplots(2,2, sharex=True, sharey=True)
ax[0][0].plot(x,y)
ax[0][0].set_xlabel("firstx")
ax[1][0].set_title("Second")
```

```python
fig, ax = plt.subplots()
```

```python
fig3 = plt.figure(figsize=(8,5))
# Creating main axes - values b/w 0 & 1
axes_main = fig3.add_axes([0.1, 0.1, 0.9, 0.9])
# # Creating inset axes - values b/w 0 & 1
axes_inset = fig3.add_axes([0.2, 0.7, 0.2, 0.2])
axes_inset.plot(x,y, 'r--')
axes_main.plot(y,x, 'r--')
```
