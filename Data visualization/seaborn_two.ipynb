```python
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
# to avoid unnecessary warnings
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
print(tips.shape, iris.shape)
```

```python
tips.head(3)
```

```python
# Axes level plot
# stipplot , between a numerical data and categorical data
sns.stripplot(data=tips, x=tips.day, y=tips.total_bill, hue=tips.time)
```

```python
# Figure lebel strip plot
sns.catplot(data=tips, x=tips.day, y=tips.total_bill, kind='strip')
```

```python
# Axis level plot
# swarm plot , between a numerical data and categorical data
# for small data
sns.swarmplot(data=tips, x=tips.day, y=tips.total_bill, hue=tips.time)
```

```python
# Figure level swarm plot
sns.catplot(data=tips, x=tips.day, y=tips.total_bill, kind='swarm')
```

```python
# Axis level plot
# box plot
sns.boxplot(data=tips, x=tips.day, y=tips.total_bill)
```

```python
sns.boxplot(data=tips, x=tips.sex, y=tips.total_bill)
```

```python
# Figure level box plot
sns.catplot(data=tips,x='day',y='total_bill', kind='box')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex')
```

# Estimate plot

```python
sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
# Figure level bar plot
sns.catplot(data=tips, x='sex', y='total_bill', kind='bar',hue='smoker')
```

```python
sns.catplot(data=tips, x='sex', y='total_bill', estimator=np.std, kind='bar')
```

```python
sns.countplot(data=tips, x='smoker')
```

```python
sns.countplot(data = tips, x='smoker', hue='sex')
```

# Reg plot

```python
# scatterplot with regression line or prediction line
sns.regplot(data=tips, x='total_bill', y='tip')
```

```python
# axis plot and with hue
sns.lmplot(data=tips, x='total_bill', y='tip', hue='day')
```

```python
# Residual of Regression or error of linearRegression
sns.residplot(data=tips,x='total_bill',y='tip')
```
