# 2. Distribution Plot

<h2>Types of Distribution plot</h2>
<h3><ul>
    <li>histogram Plot : sns.histplot(), Axes level plot</li>
    <li>Karnel Density Estimate Plot : sns.kdeplot(), Axes level plot</li>
    <li>rug Plot : sns.rugplot(), Axes level plot</li>
    <li>displot plot : sns.rugplot(), FIgure level plot.(can do all(hist,kde,rug) by kind='hist'/'kde'/'rug' also can overlay by kde=True,rug=True)</li>
    
</ul></h3>

```python
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='hist', bins=20, hue='sex')
```

```python
sns.displot(data=tips, x='total_bill', bins=20, hue='sex', element='step', kind='hist')
```

# Overlay histogram and kdePlot

```python
sns.displot(data=tips, x='total_bill', kind='hist', kde=True)
```

```python
tips.head()
```

```python
sns.displot(data=tips, x='total_bill', y='tip')
```

```python
sns.displot(data=tips, x='total_bill', y='tip', kind='kde', rug=True)
```

```python
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True)
```

# Now let's check whether iris datas are in Normal Distribution or not

```python
print(iris.shape)
iris.head()
```

```python
sns.displot(data=iris, x='sepal_length', kde=True)
```

```python
from scipy import stats
```

```python
stats.probplot(iris.sepal_length, plot=plt)
plt.show()
```

# the more the data point overlap the straight line the more the datas is normally distributed
